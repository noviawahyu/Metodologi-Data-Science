Library PANDAS supaya bisa urut

```python
# sorting besar ke kecil (kanan-kiri)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

negara = ('Argentina','Brazil','Chile','Denmark','Finlandia','German','Hungaria','Indonesia','Jepang','Kamerun')
populasi = (45380000,212600000,19120000,5831000,5531000,83240000,9750000,273500000,12580000,2655000)

# data frame
df = pd.DataFrame({'Country':negara,'Population':populasi,})
df.sort_values(by='Population',inplace = True)

# mengubah berdasar df (populasi)
kor_x = np.arange(len(df))
plt.bar(kor_x,df['Population'],tick_label = df['Country'])
plt.xticks(rotation=90) # mengubah rotasi untuk nama negara
plt.ylabel('Populasi (juta)')
plt.title('Populasi 10 Negara di dunia') # menambah judul
plt.show()
```

```python
# sorting kecil ke besar (kiri-kanan)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

negara = ('Argentina','Brazil','Chile','Denmark','Finlandia','German','Hungaria','Indonesia','Jepang','Kamerun')
populasi = (45380000,212600000,19120000,5831000,5531000,83240000,9750000,273500000,12580000,2655000)

# data frame
df = pd.DataFrame({'Country':negara,'Population':populasi,})
df.sort_values(by='Population',inplace = True, ascending = False)

# mengubah berdasar df (populasi)
kor_x = np.arange(len(df))
plt.bar(kor_x,df['Population'],tick_label = df['Country'])
plt.xticks(rotation=90) # mengubah rotasi untuk nama negara
plt.ylabel('Populasi (juta)')
plt.title('Populasi 10 Negara di dunia') # menambah judul
plt.show()
```

Highlight kasih warna

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

negara = ('Argentina','Brazil','Chile','Denmark','Finlandia','German','Hungaria','Indonesia','Jepang','Kamerun')
populasi = (45380000,212600000,19120000,5831000,5531000,83240000,9750000,273500000,12580000,2655000)

# data frame
df = pd.DataFrame({'Country':negara,'Population':populasi,})
df.sort_values(by='Population',inplace = True)

# mengubah berdasar df (populasi)
kor_x = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
warna [-3] = '#FF0000'

plt.bar(kor_x,df['Population'],tick_label = df['Country'],color = warna)
plt.xticks(rotation=90) # mengubah rotasi untuk nama negara
plt.ylabel('Populasi (juta)')
plt.title('Populasi Negara Terbesar Ketiga') # menambah judul
plt.show()
```

Memperbesar

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

negara = ('Argentina','Brazil','Chile','Denmark','Finlandia','German','Hungaria','Indonesia','Jepang','Kamerun')
populasi = (45380000,212600000,19120000,5831000,5531000,83240000,9750000,273500000,12580000,2655000)

# data frame
df = pd.DataFrame({'Country':negara,'Population':populasi,})
df.sort_values(by='Population',inplace = True)

# mengubah berdasar df (populasi)
kor_x = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
warna [-3] = '#FF0000'

plt.figure(figsize=(20,10))

plt.bar(kor_x,df['Population'],tick_label = df['Country'],color = warna)
plt.xticks(rotation=90) # mengubah rotasi untuk nama negara
plt.ylabel('Populasi (juta)')
plt.title('Populasi Negara Terbesar Ketiga') # menambah judul
plt.show()
```
