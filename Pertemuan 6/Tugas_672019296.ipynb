## No. 3

Pada soal no 3 saya menggunakan Bar Chart karena dapat menampilkan jumlah penjualan tertinggi hingga terendah secara urut dan begitu juga sebaliknya.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datapenjualan_makananpenutup = ('Donat','Pastel','Kue Coklat','Ice Cream','Puding Vanila','Brownies','Puding Strawberi',
                                'Puding Coklat','Ice Cream Nutela','Kue Coklat-Keju','Kue Wajik','Kue Sus','Mochi')
jml_jual = (14,5,12,19,6,8,12,9,10,17,2,9,13)

df = pd.DataFrame({'Penjualan':datapenjualan_makananpenutup, 'Terjual':jml_jual})
df.sort_values(by='Terjual',inplace = True, ascending = False)

kor_x = np.arange(len(df))
warna = ['#48D1CC' for _ in range(len(df))]
warna [-1] = '#C71585'
warna [-2] = '#C71585'
warna [-3] = '#C71585'

plt.figure(figsize=(16,9))
plt.bar(kor_x,df['Terjual'],tick_label = df['Penjualan'],color = warna)
plt.xticks(rotation=90)
plt.ylabel('Jumlah Terjual')
plt.xlabel('Nama Makanan')
plt.title('Daftar Penjualan Makanan Penutup') 
plt.show()
```

Kesimpulan: Makanan penutup yang paling diminati adalah ice cream, kue coklat-keju, dan donat
Untuk makanan penutup yang kurang diminati dan harus disingkarkan Kafe Biru adalah puding vanila, pastel, dan kue wajik.

Kesimpulan:
